## 1.1 - Add a polygon to the figure

```python
def generate_a_drawing(figsize, U, V, noise=0.0):
    """Generate the drawing for data U and V with an optional noise."""
    
    # Generate a figure
    fig = plt.figure(figsize=(figsize, figsize))
    ax = plt.subplot(111)
    
    # Parameter of the figure
    plt.axis('Off')
    ax.set_xlim(0, figsize)
    ax.set_ylim(0, figsize)
    
    # Fill the polygon defined by U and V
    ax.fill(U, V, "k")
    fig.canvas.draw()
    
    # Add a noise to the data
    imdata = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)[::3].astype(np.float32)
    imdata = imdata + noise * np.random.random(imdata.size)
    
    # Close the figure
    plt.close(fig)
    
    return imdata
```

## 1.2 - Add a rectangle to the figure

```python
def generate_a_rectangle(noise=0.0, free_location=False,
                         generate_a_drawing=generate_a_drawing):
    """Generate a rectangle on the figure."""
    
    # Defintion of the size of the figure
    figsize = 1.0
    
    # Initialisation of the position arrays
    U = np.zeros(4)
    V = np.zeros(4)
    
    if free_location:
        corners = np.random.random(4)
        top = max(corners[0], corners[1])
        bottom = min(corners[0], corners[1])
        left = min(corners[2], corners[3])
        right = max(corners[2], corners[3])
        
    else:
        side = (0.3 + 0.7 * np.random.random()) * figsize
        top = figsize/2 + side/2
        bottom = figsize/2 - side/2
        left = bottom
        right = top

    # Deinfition of the rectangle
    U[0] = U[1] = top
    U[2] = U[3] = bottom
    V[0] = V[3] = left
    V[1] = V[2] = right
    
    return generate_a_drawing(figsize, U, V, noise)
```

```python
im = generate_a_rectangle(10, True)
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray');
```

## 1.3 - Add a disk to the figure

```python
def generate_a_disk(noise=0.0, free_location=False, generate_a_drawing=generate_a_drawing):
    """Generate a disk on the figure."""
    
    # Defintion of the size of the figure
    figsize = 1.0
    
    # Definition of the center
    if free_location:
        center = np.random.random(2)
    else:
        center = (figsize/2, figsize/2)
    
    # Definition of the radius
    radius = (0.3 + 0.7 * np.random.random()) * figsize/2
    
    # intialisation of the point of the circle
    N = 50
    U = np.zeros(N)
    V = np.zeros(N)
    
    # Compute the positions
    for i, t in enumerate(np.linspace(0, 2*np.pi, N)):
        U[i] = center[0] + np.cos(t) * radius
        V[i] = center[1] + np.sin(t) * radius
        
    return generate_a_drawing(figsize, U, V, noise)
```

```python
im = generate_a_disk(10)
plt.imshow(im.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray');
```

## 1.4 - Add a triangle to the figure

```python
def generate_a_triangle(noise=0.0, free_location=False, generate_a_drawing=generate_a_drawing):
    """Generate a triangle on the figure."""
    
    # Defintion of the size of the figure
    figsize = 1.0
    
    # Definition of the points of the figure
    if free_location:
        U = np.random.random(3)
        V = np.random.random(3)
    else:
        size = (0.3 + 0.7 * np.random.random()) * figsize/2
        middle = figsize/2
        U = (middle, middle+size, middle-size)
        V = (middle+size, middle-size, middle-size)
        
    imdata = generate_a_drawing(figsize, U, V, noise)
    
    return [imdata, [U[0], V[0], U[1], V[1], U[2], V[2]]]
```

```python
[im, v] = generate_a_triangle(20, False)
plt.imshow(im.reshape(IMAGE_SIZE, IMAGE_SIZE), cmap='gray');
```

## 1.5 - Generate a Training Dataset and a Testing dataset for classification

```python
def generate_dataset_classification(nb_samples, noise=0.0, free_location=False):
    """Generate a dataset for the classification."""
    
    # Getting im_size:
    im_size = generate_a_rectangle().shape[0]
    X = np.zeros([nb_samples, im_size])
    Y = np.zeros(nb_samples)
    
    # Display the creation indication   
    for i in tqdm.tqdm(range(nb_samples)):

        category = np.random.randint(3)
        if category == 0:
            X[i] = generate_a_rectangle(noise, free_location)
        elif category == 1: 
            X[i] = generate_a_disk(noise, free_location)
        else:
            [X[i], V] = generate_a_triangle(noise, free_location)
            
        Y[i] = category
    
    # Add a noise
    X = (X + noise) / (255 + 2 * noise)
    
    return [X, Y]
```

```python
def generate_test_set_classification(noise=20, free_location=False):
    """Generate a testing set."""
    
    # Seed
    np.random.seed(42)
    
    # Generate the dataset
    [X_test, Y_test] = generate_dataset_classification(300, noise, free_location)
    
    # To categorical the y
    Y_test = np_utils.to_categorical(Y_test, 3) 
    
    return [X_test, Y_test]
```

## 1.7 - Creation of a training and testing dataset

```python
# Build the training set
X_train, Y_train = generate_dataset_classification(300, 20)
y_train = np_utils.to_categorical(Y_train, 3)

# Build the testing set
X_test, y_test = generate_test_set_classification(noise=0, free_location=False)
```

## 1.8 - Model

```python
# Definition of the model
model = Sequential()

# Add a layer with 3 neurons for the three output and an activation softmax
# for the classification
model.add(Dense(3, activation="softmax", input_shape=(IMAGE_SIZE**2, )))
```

# 6 - A regression problem

## 6.1 - Creating a training set and a testing set for the regression

```python
def generate_dataset_regression(nb_samples, noise=0.0):
    """Generate a dataset for the regression."""
    
    # Getting im_size:
    im_size = IMAGE_SIZE**2 #generate_a_triangle()[0].shape[0]
    
    # Initialisation of the saving array
    X = np.zeros([nb_samples, im_size])
    Y = np.zeros([nb_samples, 6])
    
    # Create the data for the regression
    for i in tqdm.tqdm(range(nb_samples)):

        [X[i], Y[i]] = generate_a_triangle(noise, True)
    
    # Add a noise
    X = (X + noise) / (255 + 2 * noise)
    
    return [X, Y]
```

```python
def generate_test_set_regression(noise=20):
    """Generate a testing set for the regression"""
    
    # Seed
    np.random.seed(42)
    
    # Generate the testing data
    [X_test, Y_test] = generate_dataset_regression(300, noise)
    
    return [X_test, Y_test]
```

## 6.2 - Visualisation of the prediction

```python
def visualize_prediction(x, y, ax=None, color='r',label="True"):
    """Visualisation of the prediction."""
    
    # Parameters of the figure
    if ax is None:
        fig, ax = plt.subplots(figsize=(5, 5))
        ax.set_xlim([0,1])
        ax.set_ylim([0,1])
    
    # reshape of x as an image
    I = x.reshape((IMAGE_SIZE, IMAGE_SIZE))
    ax.imshow(I, extent=[-0.15, 1.15, -0.15, 1.15], cmap='gray')
    
    # Add the veryices of the triangle to the figure
    xy = y.reshape(3, 2)
    tri = patches.Polygon(xy, closed=True, fill=False, edgecolor=color,
                          linewidth=2, alpha=0.5, label=label)
    ax.add_patch(tri)
    
    # Add legend
    ax.legend()
```

## 6.3 - Generating a Training and Testing set

```python
# Build a training set
X_train_reg, Y_train_reg = generate_dataset_regression(10000, 20)

# Build a testing set
X_test_reg, Y_test_reg = generate_dataset_regression(300, 20)
```

```python
# Reshape of the sets
X_train_reg = X_train_reg.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
X_test_reg = X_test_reg.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

## 6.4 - Ordre lexicographique sur les points

```python
def sortingPoints(Y):
    """Sort the points according to their x values."""
    
    # Reshape Y
    Y = Y.reshape((-1, 3, 2))
    
    # Saving array
    new_Y = np.zeros((np.shape(Y)))
    
    # Loop over each triangles
    for i, points in enumerate(Y):
        
        #Convert as a list
        points_list = list(points)
        
        # Sort the points
        points_list.sort(key=lambda x: x[0])
        
        # Put the last point as the second point
        points_list = [points_list[0], points_list[2], points_list[1]]
        
        # Update Y
        new_Y[i, :, :] = points_list
        
    # Reshape Y
    new_Y = new_Y.reshape((-1, 6))
    
    return new_Y
```

```python
# Renormalisation of the vertices
y_train_reg = sortingPoints(Y_train_reg)
y_test_reg = sortingPoints(Y_test_reg)
```

## 6.5 - Model

```python
# Definition of the model
reg_model = Sequential([
    BatchNormalization(),
    Conv2D(8, 3, activation='relu', padding="same", input_shape=(IMAGE_SIZE, IMAGE_SIZE, 1)),
    Conv2D(8, 3, activation='relu', padding="same"),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(16, 3, activation='relu', padding="same"),
    Conv2D(16, 3, activation='relu', padding="same"),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(32, 3, activation='relu', padding="same"),
    Conv2D(32, 3, activation='relu', padding="same"),
    MaxPooling2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(6, activation="relu")
])

# Compiler of the model
reg_model.compile(loss='mean_squared_error', optimizer="adam", metrics=['accuracy'])
```

## 6.6 - Fitting of the model and evaluation

```python
reg_model.fit(X_train_reg, y_train_reg, epochs=25, batch_size=32, verbose=1)
```

```python
reg_model.summary()
```

```python
# Evaluation of the model
evaluation = reg_model.evaluate(X_test_reg, y_test_reg)
loss = evaluation[0]
accuracy = round(evaluation[1] * 100, 2)
print("Loss: ", loss)
print("Accuracy: ", accuracy, "%")
```

__Results__:
\begin{itemize}
    \item Here, as our model has much more variables than the previous one, we increased the amounts of examples for avoiding overfitting. In particular, we used a dataset of 10000 shapes.
    \item For the normalisation of the points, the importance is to give an order to the points for helping the algorithm to recognise them. For that, we just sorted the points according to their abscisses x.
    \item For the architecture, we get inspired by a simple encoder which is able to extract the relevant features for our problem.
\end{itemize}

## 6.7 - Evaluation of the model

```python
def visualise_prediction_n(n, X, y_true, y_pred):
    """Display the true and predicted vertices for n the first figures in X"""

    # Display some examples
    fig, axs = plt.subplots(n//3, 3, figsize=(15, 15))

    # Display the figure
    for i in range(n):

        # Visualisation of the true vertices
        visualize_prediction(X[i], y_true[i], axs[i//3, i%3], color='g')
        
        # Visualisation of the true vertices
        visualize_prediction(X[i], y_pred[i], axs[i//3, i%3], color='r', label="Predicted")

    # Tight layout
    fig.tight_layout()
```

# 7 - Image Denoising

## 7.1 - Generate Mirror Figures

```python
def generate_mirror_drawings(figsize, U, V, noise=20):
    """Generate mirror drawings for data U and V for the given noise."""
    
    # Generate a figure
    fig = plt.figure(figsize=(figsize, figsize))
    ax = plt.subplot(111)
    
    # Parameter of the figure
    plt.axis('Off')
    ax.set_xlim(0, figsize)
    ax.set_ylim(0, figsize)
    
    # Fill the polygon defined by U and V
    ax.fill(U, V, "k")
    fig.canvas.draw()
    
    # Add a noise to the data
    imdata_2 = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)[::3].astype(np.float32)
    imdata_1 = imdata_2 + noise * np.random.random(imdata_2.size)
    
    # Close the figure
    plt.close(fig)
    
    return [imdata_1, imdata_2]
```

## 7.2 - Generate a dataset for pairs of images

```python
def generate_dataset_pairs(nb_samples, noise=40, free_location=False):
    """Generate a dataset for the classification."""
    
    # Getting im_size:
    im_size = generate_a_rectangle().shape[0]
    X = np.zeros([nb_samples, im_size])
    Y = np.zeros([nb_samples, im_size])
    
    # Display the creation indication   
    for i in tqdm.tqdm(range(nb_samples)):

        category = np.random.randint(3)
        if category == 0:
            X[i], Y[i] = generate_a_rectangle(noise, free_location, 
                                              generate_a_drawing=generate_mirror_drawings)
        elif category == 1: 
            X[i], Y[i] = generate_a_disk(noise, free_location,
                                         generate_a_drawing=generate_mirror_drawings)
        else:
            [Ims, V] = generate_a_triangle(noise, free_location,
                                           generate_a_drawing=generate_mirror_drawings)
            X[i] = Ims[0]
            Y[i] = Ims[1]
    
    return [X, Y]
```

## 7.3 - Function for generating a testing set of paris of images

```python
def generate_test_set_pairs(noise=40, free_location=False):
    """Generate a testing set."""
    
    # Seed
    np.random.seed(42)
    
    # Generate the dataset
    [X_test, Y_test] = generate_dataset_pairs(300, noise, free_location)
    
    return [X_test, Y_test]
```

## 7.4 - Generate a Testing set and a training set

```python
# Build a training set
X_train_noisy, Y_train_noisy = generate_dataset_pairs(300, 500, True)
X_train_noisy = X_train_noisy.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
Y_train_noisy = Y_train_noisy.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))

# Build a testing set
X_test_noisy, Y_test_noisy = generate_test_set_pairs(500, True)
X_test_noisy = X_test_noisy.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
Y_test_noisy = Y_test_noisy.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

```python
# Normalisation of the training set
y_train_noisy = np_utils.to_categorical(Y_train_noisy / 255, 2)

# Normalisation of the testing set
y_test_noisy = np_utils.to_categorical(Y_test_noisy / 255, 2)
```

```python
# Visualisation of one result
fig, axs = plt.subplots(1, 2)
axs[0].imshow(X_train_noisy[1].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray');
axs[1].imshow(Y_train_noisy[1].reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray');
```

## 7.5 - Unet Model

```python
# For the denoising model, we get inspired by the U_net model
U_net = Sequential([
    BatchNormalization(),
    Conv2D(4, 3, padding="same", activation="relu", 
           input_shape=(IMAGE_SIZE, IMAGE_SIZE, 1)),
    Conv2D(4, 3, padding="same", activation="relu"),
    MaxPooling2D(pool_size=(2, 2)),
    
    Conv2D(8, 3, padding="same", activation="relu"),
    Conv2D(8, 3, padding="same", activation="relu"),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    UpSampling2D(size=(2,2), interpolation="bilinear"),
    Conv2D(4, 2, padding="same", activation="relu"),
    Conv2D(4, 2, padding="same", activation="relu"),
    
    UpSampling2D(size=(2,2), interpolation="bilinear"),
    Conv2D(2, 2, padding="same", activation="relu"),
    Conv2D(2, 2, padding="same", activation="relu"),
    BatchNormalization(),
    
    Conv2D(2, 1, padding="same", activation="softmax"),
])

# Compiler of our model
U_net.compile(optimizer='adam', loss='categorical_crossentropy')
```

## 7.6 - Fitting of the model

```python
U_net.fit(X_train_noisy, y_train_noisy, epochs=50, verbose=1)
```

## 7.7 - Evaluation of the model

```python
# Evaluation of the model
evaluation = U_net.evaluate(X_test_noisy, y_test_noisy)
loss = evaluation
print("Loss: ", loss)
```
